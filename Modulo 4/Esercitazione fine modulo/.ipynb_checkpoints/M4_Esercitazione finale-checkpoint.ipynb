# Analisi diffusione COVID-19 nel mondo

## Richieste:

    - Verificare le dimensioni del dataset e le diciture presenti nell'intestazione
    - Per ogni continente trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di non considerare eventuali locazioni che nel dataset non appartengono a nessun continente.
    - Sviluppare una funzione che prenda in input il dataset e due nomi di continenti, e che ne confronti i seguenti relativi descrittori: minimo, massimo, media, percentuale rispetto al numero totale dei casi nel mondo
    - Stesso tipo di analisi per i vaccini
    - Confronto tra Europa, Sud America e Oceania

```python
#considerando la richiesta, ho deciso di creare un dataset di dimensione ridotta che contenga solo le colonne potenzialmente utili ai fine dell'analisi richiesta
df2 = df.loc[:,["continent","location", "date", 
                "total_cases", "new_cases", "total_deaths", "new_deaths", "total_vaccinations",
                "people_vaccinated", "people_fully_vaccinated", "total_boosters","new_vaccinations", 
                "population"]]
#dataframe in cui sono state rimosse le righe che non riportavano valori nella colonna continenti
dfnotnull= df2[df2.continent.notnull()]
dfnotnull
```

#### Numero di casi per continente

```python
#Usata la colonna new_cases in quanto indica il dato "nuovi casi" per data, la colonna total_cases sembrerebbe invece un campo cumulativo.

continenti = dfnotnull.groupby("continent")
continenti["new_cases"].sum()
```

### Comparazione tra continenti

```python
def comparazione_casi(df,continente1,continente2):
    """
    La funzione permette di mettere a confronto la diffusione del COVID-19 tra due continenti scelti dall'utente
    
    """
    casi_totali_mondo = df.new_cases.sum()
    continenti = set(df.continent)
    
    for continente in continenti:
        condizione1 = df.continent == continente1
        condizione2 = df.continent == continente2
        percentuale_vs_mondo1 = round(df.loc[condizione1, "new_cases"].sum()/casi_totali_mondo*100,2)
        percentuale_vs_mondo2 = round(df.loc[condizione2, "new_cases"].sum()/casi_totali_mondo*100,2)
        casi_giornalieri_medi1 = round(df.loc[condizione1, "new_cases"].mean(),2)
        casi_giornalieri_medi2 = round(df.loc[condizione2, "new_cases"].mean(),2)
        casi_minimi1 = df.loc[condizione1, "new_cases"].min()
        casi_minimi2 = df.loc[condizione2, "new_cases"].min()
        casi_massimi1 = df.loc[condizione1, "new_cases"].max()
        casi_massimi2 = df.loc[condizione2, "new_cases"].max()
    data = {
        "% casi vs mondo": [percentuale_vs_mondo1, percentuale_vs_mondo2],
        "casi minimi": [casi_minimi1, casi_minimi2],
        "casi massimi": [casi_massimi1, casi_massimi2],
        "casi giornalieri medi":[casi_giornalieri_medi1, casi_giornalieri_medi2]}
    risultato = pd.DataFrame(data, index=[continente1, continente2])
    return risultato    
```

```python
#dataframe senza i valori nulli sotto continente: dato che le location che non ripotavano il continente corrispondevano a dati cumulativi o identificativi non meglio specificati
comparazione_casi(dfnotnull,"Africa","South America")
```

```python
#dataframe che include i valori nulli
comparazione_casi(df2,"Africa", "South America")
```

```python
def comparazione_vaccini(df,continente1,continente2):
    """
    La funzione permette di mettere a confronto le vaccinazioni contro il COVID-19 tra due continenti scelti dall'utente
    
    """
    vaccini_totali_mondo = df.new_vaccinations.sum()
    continenti = set(df.continent)
    
    for continente in continenti:
        condizione1 = df.continent == continente1
        condizione2 = df.continent == continente2
        percentuale_vs_mondo1 = round(df.loc[condizione1, "new_vaccinations"].sum()/vaccini_totali_mondo*100,2)
        percentuale_vs_mondo2 = round(df.loc[condizione2, "new_vaccinations"].sum()/vaccini_totali_mondo*100,2)
        vaccini_giornalieri_medi1 = round(df.loc[condizione1, "new_vaccinations"].mean(),2)
        vaccini_giornalieri_medi2 = round(df.loc[condizione2, "new_vaccinations"].mean(),2)
        vaccini_minimi1 = df.loc[condizione1,"new_vaccinations"].min()
        vaccini_minimi2 = df.loc[condizione2,"new_vaccinations"].min()
        vaccini_massimi1 = df.loc[condizione1,"new_vaccinations"].max()
        vaccini_massimi2 = df.loc[condizione2,"new_vaccinations"].max()
    data = {
        "% vaccini vs mondo": [percentuale_vs_mondo1, percentuale_vs_mondo2],
        "valore minimo": [vaccini_minimi1, vaccini_minimi2],
        "valore massimo": [vaccini_massimi1, vaccini_massimi2],
        "vaccini giornalieri medi":[vaccini_giornalieri_medi1, vaccini_giornalieri_medi2]}
    risultato = pd.DataFrame(data, index=[continente1, continente2])
    return risultato    
```

```python
comparazione_vaccini(dfnotnull,"Europe","Asia")
```

## Differenze tra Europa, Sud America, Oceania

```python
casi_totali_mondo = dfnotnull.new_cases.sum()
vaccini_totali_mondo = dfnotnull.new_vaccinations.sum()
continenti = set(df.continent)
    
for continente in continenti:
    condizione1 = dfnotnull.continent == "Europe"
    condizione2 = dfnotnull.continent == "South America"
    condizione3 = dfnotnull.continent == "Oceania"
    percentuale_vs_mondo1 = round(dfnotnull.loc[condizione1, "new_cases"].sum()/casi_totali_mondo*100,2)
    percentuale_vs_mondo2 = round(dfnotnull.loc[condizione2, "new_cases"].sum()/casi_totali_mondo*100,2)
    percentuale_vs_mondo3 = round(dfnotnull.loc[condizione3, "new_cases"].sum()/casi_totali_mondo*100,2)
    casi_giornalieri_medi1 = round(dfnotnull.loc[condizione1, "new_cases"].mean(),2)
    casi_giornalieri_medi2 = round(dfnotnull.loc[condizione2, "new_cases"].mean(),2)
    casi_giornalieri_medi3 = round(dfnotnull.loc[condizione3, "new_cases"].mean(),2)
    vaccini_vs_mondo1 = round(dfnotnull.loc[condizione1, "new_vaccinations"].sum()/vaccini_totali_mondo*100,2)
    vaccini_vs_mondo2 = round(dfnotnull.loc[condizione2, "new_vaccinations"].sum()/vaccini_totali_mondo*100,2)
    vaccini_vs_mondo3 = round(dfnotnull.loc[condizione3, "new_vaccinations"].sum()/vaccini_totali_mondo*100,2)
    vaccini_giornalieri_medi1 = round(dfnotnull.loc[condizione1, "new_vaccinations"].mean(),2)
    vaccini_giornalieri_medi2 = round(dfnotnull.loc[condizione2, "new_vaccinations"].mean(),2)
    vaccini_giornalieri_medi3 = round(dfnotnull.loc[condizione3, "new_vaccinations"].mean(),2)
    
    
    
data = {
    "% casi vs mondo": [percentuale_vs_mondo1, percentuale_vs_mondo2, percentuale_vs_mondo3],
    "casi giornalieri medi":[casi_giornalieri_medi1, casi_giornalieri_medi2,casi_giornalieri_medi3],
    "% vaccini vs mondo": [vaccini_vs_mondo1,vaccini_vs_mondo2,vaccini_vs_mondo3],
    "vaccini giornalieri medi": [vaccini_giornalieri_medi1,vaccini_giornalieri_medi2,vaccini_giornalieri_medi3]}

risultato = pd.DataFrame(data, index=["Europe", "South America", "Oceania"])
display(risultato)
```

Nel confronto tra i tre continenti l'Europa detiene la percentuale più alta di casi covid rispetto al totale dei casi mondo.
I casi medi giornalieri tra Europa e Sud America differiscono solo lievemente, mentre l'Oceania ha un valore decisamente più basso.
Per quanto riguarda i vaccini, il Sud America ha una media giornaliera molto più alta, mentre Europa e Oceania sono allineate intorno al 60000 al giorno.
